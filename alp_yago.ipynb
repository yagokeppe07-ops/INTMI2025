##Operadores
Adição +

Subtração -

Divisão /

Multiplicação *

Resto da divisão %

Divisão inteira //

Exponenciação **

```python
numero1 = 10
numero2 = 5

print("a soma e igual a:", numero1 + numero2)
print("a subtração e igual a:", numero1 - numero2)
print("a divisão e igual a:", numero1 / numero2)
print("a multiplicação e igual a:", numero1 * numero2)
print("resto da divisão e igual a:", numero1 % numero2)
print("a divisão inteira e igual a:", numero1 // numero2)
print("a exponencial e igual a:", numero1 ** numero2)
```

##Estrutura Condicional

if - se

elif - senão se

else - senão

```python
nota = float(input("digite a sua nota"))

if(nota >=7):
  print("aprovado")
elif (nota>=5):
    print("recuperação")
else:
  print("reprovado")
```

- Crie um algoritimo que leia a sua idade e dependendo da idade mostre as frases na tela:

até 12 anos você é uma criança

13 - 18  - você é adolescente

19 - 60  - você é adulto

acima de 60 - você é idoso

```python
idade = int(input("digite sua idade"))

if(idade <=12):
  print("você é uma criança")
elif(idade >=13 and idade <=18):
  print("você é adolescente")
elif(idade >=19 and idade <=60):
  print("você é adulto")
```
